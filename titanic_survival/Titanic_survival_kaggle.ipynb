```python
import numpy as np
import pandas as pd
```

```python
df_train= pd.read_csv('train.csv')
df_test=pd.read_csv('test.csv')
df_train=df_train.copy()
df_test=df_test.copy()
```

```python
#The Dataset Cleaner(Training):
df_train= pd.read_csv('train.csv')
df_train['Cabin'] = df_train['Cabin'].str[0].fillna('U') ##Keeps only the first letter of the Cabin details,it seems
# to be that instead of tring to identify which cabins they belonged to,
#people just make a new label Unknown.
df_train=df_train[df_train['Cabin'] != 'T'] #Only one person from that level, can be considered a outlier
df_train = df_train.drop('Ticket', axis=1) #Seems pretty useless at the moment, maybe when im more skilled i can do something with it
df_train['Title'] = df_train['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False) # main reason we can identify titles, understsnd the
#function well for future use
df_train['Title']= df_train['Title'].replace(['Dr','Rev','Major','Major','Col','Countess','Capt','Sir','Lady','Don','Jonkheer'],'Rare')
df_train['Title']=df_train['Title'].replace('Mlle','Miss')
df_train['Title']=df_train['Title'].replace('Ms','Miss')
df_train['Title']=df_train['Title'].replace('Mme','Mrs')
df_train=df_train.drop('Name',axis=1)
#Above code is done to get rid of outliers, and group unique values into a certain category, should help in getting a pattern
df_train['Sex'] = df_train['Sex'].apply(lambda x: 0 if x == 'female' else 1) #Should use either .apply() or .map() as it does vectorization
df_train=df_train[df_train['Embarked'].isnull() != True] #currently seems like the best thing to do, doesnt affect the dataset much since
#only 2 rows
df_train.loc[:,'Age'] = df_train['Age'].fillna(df_train.groupby('Title')['Age'].transform('median')) #fixes the age null issue by filling
# with median as per values in title, much better than hardcoding a generalized value. note: use before encoding
cols_to_encode = ['Title', 'Embarked', 'Cabin'] # selecting columns to encode. since its not binary like sex, we did one hot encoding
df_train = pd.get_dummies(df_train, columns=cols_to_encode, drop_first=True) # does the one hot encoding using inbuilt pandas
df_train.info()
```

```python
#The Dataset Cleaner(Testing):
df_test=pd.read_csv('test.csv')
df_test['Cabin']=df_test['Cabin'].str[0].fillna('U')
df_test=df_test.drop('Ticket',axis=1)
df_test['Title']=df_test['Name'].str.extract(r' ([A-Za-z]+)\.',expand=False)
df_test['Title']= df_test['Title'].replace(['Dr','Rev','Major','Major','Col','Countess','Capt','Sir','Lady','Don','Jonkheer'],'Rare')
df_test['Title']=df_test['Title'].replace('Mlle','Miss')
df_test['Title']=df_test['Title'].replace('Ms','Miss')
df_test['Title']=df_test['Title'].replace('Mme','Mrs')
df_test=df_test.drop('Name',axis=1)
df_test['Sex']=df_test['Sex'].apply(lambda x: 0 if x== 'female' else 1)
df_test.loc[:,'Age']=df_test['Age'].fillna(df_test.groupby('Title')['Age'].transform('median'))
df_test.loc[:,'Fare']=df_test['Fare'].fillna(df_test.groupby('Pclass')['Fare'].transform('median'))
cols_to_encode = ['Title', 'Embarked', 'Cabin']
df_test = pd.get_dummies(df_test, columns=cols_to_encode, drop_first=True) # does the one hot encoding using inbuilt pandas
model_features = df_train.drop('Survived', axis=1).columns
df_test = df_test.reindex(columns=model_features, fill_value=0)
df_test.info()
```

```python
#Initial: Always drop survived column as it would leak the data and achieve 100% accuracy just by looking at that column
Y = df_train['Survived']
X = df_train.drop('Survived', axis=1)
X_test = df_test
```

```python
#importing the models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) # for learning phase
X_test_scaled = scaler.transform(X_test) # for testing phase
```

```python
logistic_regression = LogisticRegression(max_iter=1000)
# cv=5 splits the data into 5 pieces automatically
log_scores = cross_val_score(logistic_regression, X_scaled, Y, cv=5)

print("--- Logistic Regression ---")
print(f"Scores: {log_scores}")
print(f"Average: {np.mean(log_scores):.2%}")
```

```python
random_forest = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
rf_scores = cross_val_score(random_forest, X, Y, cv=5) #cv is like running the model 5 times and averaging the accuracy

print("\n--- Random Forest ---")
print(f"Scores: {rf_scores}")
print(f"Average: {np.mean(rf_scores):.2%}")
```

```python
svm_clf = SVC(gamma='auto') #gamma tells how much influence a single training example has
svm_scores = cross_val_score(svm_clf, X_scaled, Y, cv=5)

print("\n--- SVM ---")
print(f"Scores: {svm_scores}")
print(f"Average: {np.mean(svm_scores):.2%}")
```

```python
#submission: we went with random forest as the CV score is more stable
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, Y)
final_prediction= model.predict(X_test)
submission = pd.DataFrame({
    "PassengerId": pd.read_csv('test.csv')['PassengerId'],
    "Survived": final_prediction
})

#Export
submission.to_csv('submission_stable_random_forest.csv', index=False) #index has to be false of kaggle will reject
print("Submission file created successfully!")
```
